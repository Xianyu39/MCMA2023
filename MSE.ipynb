```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
file_path = 'Normalized_table.xlsx'
df = pd.read_excel(file_path)
```

```python
row_data = df.iloc[1, 6:13]
p_data = df.iloc[1,16:23]
print(p_data)
```

```python
MSE = []
for i in range(355):
    row_data = df.iloc[i, 6:13]
    pre_data = df.iloc[i,16:23]
    temp = 0
    for j in range(len(row_data)):
        temp+=(row_data[j]-pre_data[j])*(row_data[j]-pre_data[j])
    MSE.append(temp/70000)
print(MSE)
```

```python
ave = np.average(MSE)
std = np.std(MSE)
print(ave)
print(std)
```
