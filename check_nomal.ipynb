```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
file_path = 'Problem_C_Data_Wordle.xlsx'
df = pd.read_excel(file_path)

mean_colum='Means'
means = df[mean_colum]

std_colum='Stddev'
stds = df[std_colum]

```

```python
# 创建一个范围在-4到4之间的一组数据点，用于绘制正态分布的曲线
x = np.linspace(0, 8, 100)

# 计算标准正态分布在这些数据点上的概率密度值
y = stats.norm.pdf(x, means[0], stds[0])

# 绘制正态分布曲线
plt.plot(x, y)
plt.title('Standard Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.show()
```

```python
import scipy.stats as stats
from scipy.integrate import quad
```

```python
mu, sigma = 0, 1

# 定义计算正态分布在指定区间内的概率密度的函数
def integrand(x):
    return stats.norm.pdf(x, mu, sigma)

# 计算2.5到3.5之间的概率密度
p, _ = quad(integrand, 2.5, 3.5)

# 绘制柱状图
plt.bar([1], [p], width=0.5)
plt.xticks([1], ['2.5-3.5'])
plt.ylabel('Probability Density')
plt.show()
```

```python
# 正态分布的均值和标准差
mu, sigma = means[0], stds[0]
row_data = df.iloc[0, 6:13]
for i in range(7):
    row_data[i]=row_data[i]/100
# 定义计算正态分布在指定区间内的概率密度的函数
def integrand(x):
    return stats.norm.pdf(x, mu, sigma)

# 要计算的多个区间
intervals = [(0.5, 1.5), (1.5, 2.5), (2.5, 3.5), (3.5, 4.5), (4.5, 5.5), (5.5, 6.5), (6.5, 7.5)]

# 计算每个区间的概率密度
probabilities = []
for interval in intervals:
    p, _ = quad(integrand, interval[0], interval[1])
    probabilities.append(p)

# 绘制柱状图

fig, ax = plt.subplots()
ax.bar(np.arange(len(intervals))-0.2, probabilities, width=0.4, label='Distribution 1')
ax.bar(np.arange(len(intervals))+0.2, row_data, width=0.4, label='Actual')
ax.set_xticks(np.arange(len(intervals)))
ax.set_xticklabels(['{}-{}'.format(int(interval[0]), int(interval[1])) for interval in intervals])
ax.set_ylabel('Probability Density')
ax.legend()
plt.show()
```

```python
probabilities = []
for i in range(355):
    # 正态分布的均值和标准差
    mu, sigma = means[i], stds[i]
    row_data = df.iloc[i, 6:13]
    for j in range(7):
        row_data[j]=row_data[j]/100
    # 定义计算正态分布在指定区间内的概率密度的函数
    def integrand(x):
        return stats.norm.pdf(x, mu, sigma)

    # 要计算的多个区间
    intervals = [(0.5, 1.5), (1.5, 2.5), (2.5, 3.5), (3.5, 4.5), (4.5, 5.5), (5.5, 6.5), (6.5, 7.5)]

    # 计算每个区间的概率密度
    
    for interval in intervals:
        p, _ = quad(integrand, interval[0], interval[1])
        probabilities.append(p)
matrix = np.array(probabilities).reshape(355,7)
print(matrix)
```
